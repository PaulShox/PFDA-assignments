```python
# Import Libraries
import random
import matplotlib.pyplot as plt

# Lists where difference values between the top values of the 3 dice and 2 dice rolls, and between the second highest values, will be appended.
highest_diff = []
second_highest_diff = []

# Function to roll the three dice and the two dice, and sort results in descending order. Sorting in descending order will allow highest values to line up against each other.
# https://stackoverflow.com/questions/33069476/simulating-rolling-2-dice-in-python
# https://stackoverflow.com/questions/17368554/python-how-to-simulate-the-roll-of-a-dice-100-times
# https://medium.com/@awaleedpk/roll-the-dice-build-a-fun-dice-rolling-simulator-in-python-for-beginners-ccb706bbc503

def roll_and_sort(num_dice):
    return sorted([random.randint(1,6) for _ in range(num_dice)], reverse=True) # 'reverse=True' sorts results from highest to lowest

# Run the dice rolls using a for loop # https://datagy.io/python-for-loop/
for _ in range(1000): # 1000 dice rolls
    attacker_rolls = roll_and_sort(3) # Roll and sort three dice for attacker
    defender_rolls = roll_and_sort(2) # Roll and sort two dice for defender

    # Calculate the differences
    # 3 dice highest v 2 dice highest. [0] relates to the highest die number in the sorted lists. 
    # If the attacker_roll amount is greater than the defender_roll, then it's a positive and the attacker wins.
    # If they're even, it's a zero and defender wins.
    # If the defender_roll amount is higher, it's a negative and defender wins.
    highest_diff.append(attacker_rolls[0] - defender_rolls[0])
    # 3 dice second highest v 2 dice second highest. [0] relates to the second highest die number in the sorted list
    # Same logic as above in terms of winner v loser
    second_highest_diff.append(attacker_rolls[1] - defender_rolls[1])

# Plot the results
plt.figure(figsize=(6,10))

# Plot for first differences. 3 dice (attacker) v 2 dice (defender) highest
plt.subplot(2,1,1)
plt.hist(highest_diff, bins=range(-6, 7), edgecolor='black') # range is between -6 and 7 as these are the max differences between a roll of 6 and a roll of 1
plt.title("Distribution of Highest Differences (Positives - Attacker Win; Zeroes - Defender Win; Negatives - Defender Win)", fontweight='bold')
plt.xlabel("Difference (Attacker - Defender)")
plt.ylabel("Frequency")

# Plot for second differences. 3 Dice (Attacker) second highest vs. 2 Dice (Defender) second highest
plt.subplot(2, 1, 2)
plt.hist(second_highest_diff, bins=range(-6, 7), edgecolor='black')
plt.title("Distribution of Second Highest Differences (Positives - Attacker Win; Zeroes - Defender Win; Negatives - Defender Win)", fontweight='bold')
plt.xlabel("Difference (Attacker - Defender)")
plt.ylabel("Frequency")

# Display the plots
plt.tight_layout()
plt.show()
```
