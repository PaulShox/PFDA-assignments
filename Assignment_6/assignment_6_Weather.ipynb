## **1. Import Libraries and Data**

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('hly4935.csv', skiprows=23)
data.head(5)
```

```python
data.drop(0, inplace=True)
data.head(3)
```

## **2. Check Data Types and convert where necessary**

```python
data.dtypes
```

```python
# Convert date column data type from object to datetime
data['date'] = pd.to_datetime(data['date'])
```

```python
data.dtypes
```

## **3. Make the Date Column the Index Column**

```python
data.set_index('date', inplace=True)
data.head(3)
```

## **4. Plot Temperature**

### *Using Entire Dataset*

```python
# Plot Temperature
fig = plt.subplots(figsize=(20, 5))
plt.style.use('ggplot')
sns.lineplot(data['temp']).set_title('Temperature 1996-2024', fontsize=20, fontweight='bold')
```

### *Using data from 3 days*

```python
datefrom = '2019/01/01 00:00'
dateto = '2019/01/03 23:00'
fig = plt.subplots(figsize=(20, 5))
plt.style.use('ggplot')
sns.lineplot(data=data.loc[datefrom:dateto], x='date', y='temp').set_title('Temperature 1st - 3rd January 2019', fontsize=20, fontweight='bold')
```

## **5. Plot Average Daily Temperature**

### *Using Entire Datase*

```python
# Plot Average Daily Temperature
avg_d = data['temp'].resample('1d').mean()

fig = plt.subplots(figsize=(20, 5))
plt.style.use('ggplot')
sns.lineplot(avg_d).set_title('Average Daily Temperature', fontsize=20, fontweight='bold')
```

### *Using Data from 1 Month*

```python
avg_d = data['temp'].resample('1d').mean()
datefrom = '2019/01/01'
dateto = '2019/01/31'
daily_avg = avg_d.loc[datefrom:dateto]
fig = plt.subplots(figsize=(20, 5))
plt.style.use('ggplot')
sns.lineplot(daily_avg).set_title('Average Daily Temperature January 2019', fontsize=20, fontweight='bold')
```

## **6. Plot Average Monthly Temperature**

### *Using Entire Dataset*

```python
# Plot Average Monthly Temperature
avg_m = data['temp'].resample('1m').mean()

fig = plt.subplots(figsize=(20, 5))
plt.style.use('ggplot')
sns.lineplot(avg_m).set_title('Average Monthly Temperature', fontsize=20, fontweight='bold')
```

### *Using data from 1 Year*

```python
avg_m = data['temp'].resample('1m').mean()
date_from = '2019/01/01'
date_to = '2019/12/31'
monthly_avg = avg_m[date_from:date_to]
fig = plt.subplots(figsize=(20, 5))
plt.style.use('ggplot')
sns.lineplot(monthly_avg).set_title('Average Monthly Temperature 2019', fontsize=20, fontweight='bold')
```

## **7. Plot Windspeed**

*Check data type of windspeed column*

```python
data['wdsp'].dtype
```

*Data type 'O' means there are mixed data types in the column*

*This means various codes are required to find Zero, NaN, Null, and Blank Values*

```python
data['wdsp'].isnull().sum()
```

```python
data['wdsp'].isna().sum()
```

```python
data['wdsp'].value_counts().get('0')
```

```python
data['wdsp'].value_counts().get(0)
```

```python
# https://stackoverflow.com/questions/49677060/pandas-count-empty-strings-in-a-column
(data['wdsp'].values == ' ').sum()
```

*Per the above, the windspeed column has 407 zero values, and 50 blank values*

*The blank string cells pose a problem as they won't allow the column to be converted to the int data type*

*Therefore they will be replaced with the number 0*

```python
data['wdsp'] = data['wdsp'].replace(' ', 0)
```

```python
(data['wdsp'].values == ' ').sum()
```

*It can now be seen that the number of int 0 values has gone from 128 to 178*

```python
data['wdsp'].value_counts().get(0)
```

*Convert windspeed column data type to int*

```python
# https://www.geeksforgeeks.org/convert-a-dataframe-column-to-integer-in-pandas/
data['wdsp'] = data['wdsp'].map(int)
```

```python
data['wdsp'].dtype
```

*Replace all zero values with average value*

```python
# https://stackoverflow.com/questions/69921516/replace-value-with-the-average-of-its-column-with-pandas
avg = data['wdsp'].mean()
data['wdsp'] = data['wdsp'].apply(lambda v: avg if v==0 else v)
```

*Check zero values have been replaced*

```python
data['wdsp'].value_counts().get(0)
```

```python
data['wdsp'].head(10)
```

*Plot the windspeed - using data from 1 year only*

```python
date_From = '2019/01/01'
date_To = '2019/12/31'

fig = plt.subplots(figsize=(20, 5))
plt.style.use('ggplot')
sns.lineplot(data=data.loc[date_From:date_To], x='date', y='wdsp').set_title('Windspeed 2019', fontsize=20, fontweight='bold')
```

## **8. Plot One Day Rolling Windspeed Average**

```python
avg_roll = data['wdsp'].rolling('1d').mean()
d_from = '2019/01/01'
d_to = '2019/12/31'
roll_avg = avg_roll[d_from:d_to]
fig = plt.subplots(figsize=(20, 5))
plt.style.use('ggplot')
sns.lineplot(roll_avg).set_title('1 Day Rolling Windspeed Average - 2019', fontsize=20, fontweight='bold')
```

## **9. Plot Maximum Wind Speed Each Day**

*Using data from 1 Year*

```python
wind_max = data['wdsp'].resample('1d').max()
dt_frm = '2019/01/01'
dt_to = '2019/12/31'
max_wind = wind_max[dt_frm:dt_to]
fig = plt.subplots(figsize=(20, 5))
plt.style.use('ggplot')
sns.lineplot(max_wind).set_title('Daily Maximum Wind Speed - 2019', fontsize=20, fontweight='bold')
```
