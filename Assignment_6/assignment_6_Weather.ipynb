```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('hly4935.csv', skiprows=23)
data.head(5)
```

```python
data.drop(0, inplace=True)
data.head(3)
```

```python
data.dtypes
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data.dtypes
```

```python
data.set_index('date', inplace=True)
data.head(3)
```

```python
fig = plt.subplots(figsize=(20, 5))
sns.lineplot(data['temp']).set_title('Temperature', fontsize=20, fontweight='bold')
```

```python
avg_d = data['temp'].resample('1d').mean()

fig = plt.subplots(figsize=(20, 5))
sns.lineplot(avg_d).set_title('Average Daily Temperature')
```

```python
avg_m = data['temp'].resample('1m').mean()

fig = plt.subplots(figsize=(20, 5))
sns.lineplot(avg_m).set_title('Average Monthly Temperature')
```
